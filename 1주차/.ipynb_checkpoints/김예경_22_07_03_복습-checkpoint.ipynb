```python
# DF 복사
df_csv_1 = df_csv.copy()
```

```python
# 변수명 바꾸기
df_csv_1 = df_csv_1.rename(columns={'math':'korean'})
df_csv_1 
```

```python
# 새로운 칼럼 만들기 (실수형으로)
df_csv_1['mean'] = (df_csv_1['korean']+df_csv_1['english']+df_csv_1['science'])/3
df_csv_1
```

```python
#히스토그램 그래프
df_csv_1['mean'].plot.hist()
```

```python
#특정 열의 평균 구하기
df_csv_1['science'].mean()
```

```python
df_csv_1['mean'].describe()
```

```python
import numpy as np
```

```python
df_csv_1['Good/Bad'] = np.where(df_csv_1['mean']>69,'Good','Bad')
df_csv_1
```
