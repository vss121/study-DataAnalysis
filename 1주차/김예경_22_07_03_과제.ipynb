# 과제
- 통합 연비가 가장 좋은 차는 무엇일까?
- 통합 연비가 가장 안 좋은 차는 무엇일까?

```python
import pandas as pd
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
mpg['통합 연비'] = (mpg['cty']+mpg['hwy'])/2
```

```python
mpg['통합 연비'].describe()
```

```python
print("통합 연비가 가장 안 좋은 차")
print(min(mpg['통합 연비']))
df_1 = mpg[(mpg['통합 연비']==10.5)]
df_1.head()
```

```python
print("통합 연비가 가장 좋은 차")
print(max(mpg['통합 연비']))
df_2 = mpg[(mpg['통합 연비']==39.5)]
df_2.head()
```

# 과제
- 중첩 조건문을 이용해서 다양한 파생변수를 만들기
- ex) 연비 또는 새로운 파생변수의 값에 따라 등급을 더 다양하게 나누기. 10이하 C, 10~15 B, 등

```python
import numpy as np
```

```python
mpg['통합 연비 속성']=np.where(mpg['통합 연비']>=23.5,'A', np.where(mpg['통합 연비']>=20.5,'B','C'))
mpg.head(50)
```

```python
mpg['year'].describe()
```

```python
mpg['신차 속성']=np.where(mpg['year']>=2003,'new', 'old')
mpg
```
