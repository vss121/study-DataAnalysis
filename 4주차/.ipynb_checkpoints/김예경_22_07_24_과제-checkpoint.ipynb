```python
# 변수에 값을 할당해서 iloc 추출하기
```

```python
mpg.shape[1] # 11개의 열
```

```python
# 홀수 idx
a_range= range(0,11,2)
list(a_range)
```

```python
# 짝수 idx
b_range= range(1,11,2)
list(b_range)
```

```python
mpg.iloc[5:50, a_range]
```

```python
mpg.iloc[5:50, b_range]
```
