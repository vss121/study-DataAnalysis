### mpg 데이터에 대해서 오늘 배운 내용 모두 진행
- 도시연비, 고속도로 연비
- 가장 연비가 좋은 차?
- 가장 연비가 좋지 않은 차?
- 복합연비는 어떤 차가 가장 좋은지?
- 연비가 가장 좋은 차 모델?
- 연비가 가장 나쁜 차 모델?

```python
import pandas as pd
```

```python
mpg=pd.read_csv('../mpg.csv')
```

```python
mpg
```

```python
# 도시연비, 고속도로 연비 확인해보기
mpg[['cty','hwy']]
```

```python
# 도시연비의 max값 구하기
mpg.agg(cty_max=('cty','max'))
```

```python
# 도시연비가 가장 좋은 차? -> volkswagen사의 new beetle
mpg.query('cty==35')
```

```python
# 도시연비의 min값 구하기
mpg.agg(cty_min=('cty','min'))
```

```python
# 도시연비가 가장 나쁜 차? 
# -> dodge사의 dakota pickup 4wd, durango 4wd, ram 1500 pickup 4wd, ram 1500 pickup 4wd, jeep 사의 grand cherokee 4wd
mpg.query('cty==9')
```

```python
# 고속도로 연비의 max값 구하기
mpg.agg(hwy_max=('hwy','max'))
```

```python
# 고속도로 연비가 가장 좋은 차? -> volkswagen사의 jetta와 new beetle
mpg.query('hwy==44')
```

```python
# 고속도로 연비의 min값 구하기
mpg.agg(hwy_min=('hwy','min'))
```

```python
# 고속도로 연비가 가장 나쁜 차? 
mpg.query('hwy==12')
```

### 복합 연비 만들기

```python
mpg['tot']=(mpg['cty']+mpg['hwy'])/2
```

```python
mpg
```

```python
# 복합 연비의 max값 구하기
mpg.agg(tot_max=('tot','max'))
```

```python
# 복합 연비가 가장 좋은 차? -> volkswagen사의 new beetle
mpg.query('tot==39.5')
```

```python
# 복합 연비의 min값 구하기
mpg.agg(tot_min=('tot','min'))
```

```python
# 복합 연비가 가장 나쁜 차? 
# -> dodge사의 dakota pickup 4wd, durango 4wd, ram 1500 pickup 4wd, ram 1500 pickup 4wd, jeep 사의 grand cherokee 4wd
mpg.query('tot==10.5')
```
