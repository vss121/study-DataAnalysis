# 과제1
- 다중 조건 사용해 행 추출 연습

```python
import pandas as pd
```

```python
exam = pd.read_csv('../exam.csv')
exam
```

```python
exam.query('nclass==1 | nclass==3 | nclass==5') # 1반이거나 3반이거나 5반
```

```python
exam.query('nclass==2 & math>=70') # 2반이면서 수학 점수 70이상
```

```python
exam.query('nclass==3 & english>=80') # 3반이면서 영어 점수 80이상
```

```python
exam.query('nclass==5 & science>=70') # 5반이면서 과학 점수 70이상
```

```python
exam.query('(nclass==1 | nclass==2) & math>=80') # 1반이나 2반인데 수학점수 80이상
```

```python
exam.query('math>=60 & english>=60 & science>=60') # 3과목 전부 60 이상
```

```python
exam.query('nclass!=1 & science>=80') # 1반이 아닌 반인데 과학 점수 80이상
```

```python
exam.query('math>=80 & english>=80') # 수학 80이상이고 영어 80이상
```

```python
exam.query('math < 50 | english < 50') # 수학 50미만 영어 50미만
```

```python
exam.query('(math+english+science)/3>=80') # 평균 점수 80이상
```

```python
exam.query('nclass==1 & (math+english+science)/3>=50') # 1반인데 평균 점수 50이상
```

# 과제 2
- query+[] 이용해서 여러 가지 조합

```python
exam
```

```python
exam.query('nclass==2')['math']
```

```python
exam.query('nclass==2')[['id','math']]
```

```python
exam.query('math < 50')[['id','math']]
```

```python
exam_mean = exam.assign(mean=(exam['math']+exam['english']+exam['science'])/3)[['id','mean']]
exam_mean
```

```python
exam_mean.sort_values('mean')['id']
```
