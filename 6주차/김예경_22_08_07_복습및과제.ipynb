- solid, dashed, dashdot, dotted

```python
plt.plot([1,2,3],[4,5,6], color='red',  linestyle='dashed', linewidth=3, label='line1')
plt.plot([1,2,3],[5,6,3], color='black',  linestyle='dashed', linewidth=1, label='line2')
plt.legend(loc='upper right')
```

# 2-3) 마커 설정
- https://matplotlib.org/stable/api/markers_api.html
- "." point
- o 원
- s 사각형
- D 다이아
- d 홀쭉한 다이아
- "*" star

```python
plt.plot([1,2,3],[4,5,6], marker='s', markersize=10, label='line1')
plt.plot([1,2,3],[5,6,3], marker='D', markersize=10, label='line2')
plt.legend(loc='upper right')
```

## 과제

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
df = flights.pivot('month', 'year', 'passengers')
df.head()
```

```python
ax = sns.heatmap(df)
plt.title('Heatmap of Flight by seaborn', fontsize=20)
plt.show()
```

```python
sns.heatmap(df, annot=True, fmt='d')
plt.title('Annoteat cell with numeric value', fontsize=20)
plt.show()
```

```python
sns.heatmap(df, cmap='YlGnBu')
plt.title('colormap of cmap=YlGnBu', fontsize=20)
plt.show()
```
