# 6주차 : 시각화

## 산점도 scatter

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter([1,2,3],[4,5,6]) #1
plt.scatter([4,1],[5,2]) #2

plt.show()
```

## 매직커맨드
- show() 없이 바로 시각화 가능

```python
%matplotlib inline
```

```python
plt.scatter([1,2,3],[4,5,6])
```

## 그래프 제목 / 축 이름 설정
- 한글 이름 -> 깨짐 -> 폰트 사용

```python
plt.scatter([1,2,3],[4,5,6])
plt.title('title')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

## 축 범위 설정

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(1.5,2.5)
plt.ylim(4.5,5.5)
```

## 눈금값 설정

```python
plt.scatter([1,2,3],[4,5,6])
plt.xticks([1,2,3],['A','B','C'])
```

## 범례 설정
- 위치 : best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

```python
plt.scatter([1,2,3],[4,5,6], label='first')
plt.scatter([6,5,4],[3,2,1], label='second')
plt.legend(loc='center')
```

## 그래프 크기
- figure(), figsize로

```python
plt.figure(figsize=(10,5))
plt.scatter([1,2,3],[4,5,6])
```

## 그래프 스타일

```python
import seaborn as sns
sns.set() # default = darkgrid
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style='whitegrid')
plt.scatter([1,2,3],[4,5,6])
```
