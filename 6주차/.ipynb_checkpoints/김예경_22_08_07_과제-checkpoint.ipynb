```python
# 1.그래프 스타일
- darkgrid , whitegrid , dark , white , and ticks 
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sns
sns.set() # default = darkgrid
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style='whitegrid')
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style='ticks')
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style='dark')
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style='white')
plt.scatter([1,2,3],[4,5,6])
```
